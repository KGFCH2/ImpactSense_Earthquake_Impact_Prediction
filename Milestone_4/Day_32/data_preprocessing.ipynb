```python
earth_df = df[['Latitude', 'Longitude', 'Type', 'Depth','Magnitude', 'Magnitude Type',
                'Root Mean Square','Status']]
earth_df.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Assuming 'earthquake' is your DataFrame

# Select features to predict "Root Mean Square" (drop non-numeric or encode if needed)
features = ['Latitude', 'Longitude', 'Depth', 'Magnitude']  # Use relevant numeric features

# Split data into known and missing for target column
train_df = earth_df[earth_df['Root Mean Square'].notna()]
test_df = earth_df[earth_df['Root Mean Square'].isna()]

# X and y for training
X_train = train_df[features]
y_train = train_df['Root Mean Square']

# X for prediction
X_test = test_df[features]

# Initialize and train regression model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict missing values
predicted_values = model.predict(X_test)

# Impute missing values
earth_df.loc[earth_df['Root Mean Square'].isna(), 'Root Mean Square'] = predicted_values

print(earth_df['Root Mean Square'])
```

```python
earth_df.dropna(inplace=True)
```

```python
# Feature engineering
def create_damage_potential(mag, depth):
    return 0.6 * mag + 0.2 * (700 - depth) / 700 * 10

earth_df['Damage_Potential'] = create_damage_potential(earth_df.Magnitude, earth_df.Depth)

earth_df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Select the columns to encode
columns_to_encode = ['Magnitude Type', 'Status']

# Initialize the encoder (ignore unknown categories, output dense array, integer dtype)
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False, dtype=int)

# Fit and transform the data for the selected columns
encoded_array = encoder.fit_transform(earth_df[columns_to_encode])

# Convert the encoded array to a DataFrame with proper column names
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(columns_to_encode))

# Concatenate with original DataFrame (dropping the original columns to encode)
earth_df = pd.concat([earth_df.drop(columns=columns_to_encode), encoded_df], axis=1)

# Display first few rows
earth_df.head()
```

# Summary of Data Preprocessing
